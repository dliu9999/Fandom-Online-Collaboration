```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('test.csv')[['date', 'user_id', 'username', 'tweet', 'replies_count', 'retweets_count', 'likes_count']]
```

```python
df.head()
```

```python
df.groupby('date').user_id.count().to_frame().sort_index().rename(columns={'user_id': 'num_tweets'}).plot(figsize=(10, 7), title='Number of Tweets from 5/20 to 6/3');
```

```python
df.groupby('date')[['replies_count', 'retweets_count', 'likes_count']].mean().sort_index().plot(figsize=(10, 7), title='Mean Statistics from 5/20 to 6/3');
```
